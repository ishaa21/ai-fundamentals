```python
from groq import Groq
from dotenv import load_dotenv
import os

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
print("Groq client ready")
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": "What is machine learning in one sentence?"}]
)
print(response.choices[0].message.content)
```

```python
print("Prompt tokens:", response.usage.prompt_tokens)
print("Response tokens:", response.usage.completion_tokens)
print("Total tokens:", response.usage.total_tokens)
```

```python
for temp in [0.0,0.5,1.0,2.0]:
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": "Give me a creative name for as AI startup."}],
        temperature=temp
    )
    print(f"Temperature {temp}: {response.choices[0].message.content}")
    print("-" * 50)
        
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a sarcastic assistant who answers everything with humor."},
        {"role": "user", "content": "What is Python programming?"}
    ]
)
print(response.choices[0].message.content)
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    temperature=2.0,
    messages=[{"role": "user", "content": "Who won the cricket match between India and Mars on 15th May 2025?"}]
)
print(response.choices[0].message.content)
```

```python
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": "Tell me about abc published research papers on multi-agent systems."}]
)
print(response.choices[0].message.content)
```

```python
Hallucination observation:
When asked about a real person with no training data,
the model ignored the unknown entity and generated
plausible-sounding but unverified content confidently.
This happens because LLMs predict probable tokens,
not verified facts. There is no internal fact-checker.
```
